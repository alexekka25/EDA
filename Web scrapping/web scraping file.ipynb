```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


url = 'https://www.fifaratings.com/players'

response = requests.get(url)

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing player information
table = soup.find('table')

# Extract data from the table
data = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    player_data = [column.get_text(strip=True) for column in columns]
    data.append(player_data)

# Create a DataFrame
columns = ['rank','Player', 'OVA', 'POT', 'ATT', 'SKI', 'MOV', 'POW', 'MEN', 'DEF', 'GK', 'STATS']
df = pd.DataFrame(data, columns=columns)

# Print the DataFrame
print(df)
```

```python
df
```

```python
my_dataframe = pd.DataFrame(df)

# Specify the file path where you want to save the CSV file
file_path = 'D:\data analyst\Web scraping.csv'

# Save the DataFrame to a CSV file
my_dataframe.to_csv(file_path, index=False)
```

```python
import pandas as pd
import requests
from io import StringIO

raw_url = 'https://raw.githubusercontent.com/chussboi96/Top-500-most-valuable-footballers/main/Top%20500%20Most%20Valuable%20Footballers%20in%20the%20World.csv'

# Send a GET request to the raw URL
response = requests.get(raw_url)

# Check if the request was successful
if response.status_code == 200:
   
    csv_content = StringIO(response.text)
    
   
    df = pd.read_csv(csv_content)

    print(df)
else:
    print(f"Failed to fetch the content. Status code: {response.status_code}")
```
