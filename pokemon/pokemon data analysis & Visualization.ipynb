# `Univariate analysis`

## Height and Weight

```python
custom_palette = sns.color_palette("tab20c")

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 8))

# histogram for height
sns.histplot(x='height', data=df, kde=True, color=custom_palette[4], bins=20, ax=axes[0])
axes[0].set_title('Distribution of Heights')
axes[0].set_xlabel('Height')
axes[0].set_ylabel('Frequency')
axes[0].grid(True, linestyle='--', alpha=0.7)

# histogram for weight
custom_palette = sns.color_palette("dark")
sns.histplot(x='weight', data=df, kde=True, color=custom_palette[0], bins=20, ax=axes[1])
axes[1].set_title('Distribution of Weights')
axes[1].set_xlabel('Weight')
axes[1].set_ylabel('Frequency')
axes[1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```

```python
mean_val = np.mean(df['height'])
median_val = np.median(df['height'])

print(f'Mean: {mean_val}, Median: {median_val}')
```

```python
mean_val = np.mean(df['weight'])
median_val = np.median(df['weight'])

print(f'Mean: {mean_val}, Median: {median_val}')
```

### Height is in meters
###  Weight is in Kg

# Right-Skewed Tendency: Mean Pulled by Extended Right Tail(Height)" 
 
 The mean (1.2037) is greater than the median (0.99) means distribution is right-skewed
 
 Right-skewed distributions have a longer right tail, and the mean is typically pulled in the direction of the skewness by the presence of larger values.

# Weight Distribution Analysis: Right-Skewed Tendency Revealed by Mean (46.46) Exceeding Median (30.0)"

Mean: 46.46
Median: 30.0
    
    The mean is significantly greater than the median, indicating that the distribution of the 'weight' variable is right-skewed.
    
    This suggests the presence of relatively larger values that are pulling the mean towards the right, resulting in a longer right tail in the distribution.

## type1 & type2 pokemon

```python
sns.set(style="darkgrid")

plt.figure(figsize=(12, 6))  # Adjust the width and height as needed
ax = sns.countplot(x='type1', data=df)

# Calculate percentage of each category
total = len(df['type1'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center', fontsize=10, color='black')


plt.title('Percentage of Pokemon Types')
plt.xlabel('Type1')
plt.ylabel('Percentage')

plt.show()
```

```python
sns.set(style="darkgrid")

plt.figure(figsize=(12, 6))  

ax = sns.countplot(x='type2', data=df)

# Calculate percentage of each category
total = len(df['type2'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center', fontsize=10, color='black')


plt.title('Percentage of Pokemon Types')
plt.xlabel('Type2')
plt.ylabel('Percentage')

plt.show()
```

# `Bivariate analysis`

## Top 10 Attack Pokémon

```python
top_10_attack_values = df['attack'].nlargest(10)

# Filter the DataFrame for only the top 10 attack values
df_top_10 = df[df['attack'].isin(top_10_attack_values)]

# Calculate mid-range and bottom attack values
mid_range_attack_values = df_top_10['attack'].sort_values().iloc[3:8]
bottom_attack_values = df_top_10['attack'].sort_values().iloc[:3]

# Calculate percentages
total_attack = top_10_attack_values.sum()
top_percentage = (top_10_attack_values.sum() / total_attack) * 100
mid_percentage = (mid_range_attack_values.sum() / total_attack) * 100
bottom_percentage = (bottom_attack_values.sum() / total_attack) * 100

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))

# First plot - Bar plot for top 10 attacks
sns.barplot(x='name', y='attack', data=df_top_10, hue='name', dodge=False, palette='rocket', ax=axes[0])
axes[0].set_title('Top 10 Attack Pokémon')
axes[0].set_xlabel('Name')
axes[0].set_ylabel('Attack Value')
axes[0].legend(title='Name', bbox_to_anchor=(1.05, 1), loc='upper left')

# Second plot - Donut chart for percentage distribution
axes[1].pie([top_percentage, mid_percentage, bottom_percentage],
            labels=['Top Attacks', 'Mid Attacks', 'Bottom Attacks'],
            autopct='%1.1f%%', startangle=90,
            colors=['skyblue', 'lightcoral', 'lightgreen'], wedgeprops=dict(width=0.4))
centre_circle = plt.Circle((0, 0), 0.6, color='white', fc='white', linewidth=1.25)
axes[1].add_artist(centre_circle)
axes[1].set_title('Percentage Distribution of Top, Mid, and Bottom Attacks')

plt.tight_layout()
plt.show()
```

## Top 10 Defense Pokémon

```python
top_10_attack_values = df['defense'].nlargest(10)

# Filter the DataFrame for only the top 10 attack values
df_top_10 = df[df['defense'].isin(top_10_attack_values)]

# Calculate mid-range and bottom attack values
mid_range_attack_values = df_top_10['defense'].sort_values().iloc[3:8]
bottom_attack_values = df_top_10['defense'].sort_values().iloc[:3]

# Calculate percentages
total_attack = top_10_attack_values.sum()
top_percentage = (top_10_attack_values.sum() / total_attack) * 100
mid_percentage = (mid_range_attack_values.sum() / total_attack) * 100
bottom_percentage = (bottom_attack_values.sum() / total_attack) * 100

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))

# First plot - Bar plot for top 10 attacks
sns.barplot(x='name', y='defense', data=df_top_10, hue='name', dodge=False, palette='mako', ax=axes[0])
axes[0].set_title('Top 10 Defense Pokémon')
axes[0].set_xlabel('Name')
axes[0].set_ylabel('defense Value')
axes[0].legend(title='Name', bbox_to_anchor=(1.05, 1), loc='upper left')

# Second plot - Donut chart for percentage distribution
axes[1].pie([top_percentage, mid_percentage, bottom_percentage],
            labels=['Top Defense', 'Mid Defense', 'Bottom Defense'],
            autopct='%1.1f%%', startangle=90,
            colors=['skyblue', 'lightcoral', 'lightgreen'], wedgeprops=dict(width=0.4))
centre_circle = plt.Circle((0, 0), 0.6, color='white', fc='white', linewidth=1.25)
axes[1].add_artist(centre_circle)
axes[1].set_title('Percentage Distribution of Top, Mid, and Bottom Attacks')

plt.tight_layout()
plt.show()
```

## Attack vs. Defense

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 12))

# Scatter plot
sns.scatterplot(x='attack', y='defense', data=df, ax=axes[0])
axes[0].set_title('Attack vs. Defense')
axes[0].set_xlabel('Attack')
axes[0].set_ylabel('Defense')

# Line plot
sns.lineplot(x='attack', y='defense', data=df,ax=axes[1])
axes[1].set_title('Line Plot of Attack vs. Defense')
axes[1].set_xlabel('Attack')
axes[1].set_ylabel('Defense')


corr_matrix = df[['attack', 'defense']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", ax=axes[2])
axes[2].set_title('Correlation Heatmap (Attack vs. Defense)')

# Adjust layout
plt.tight_layout()

plt.show()
```

## 1)Pattern
 ### scatter plot and lineplot:
shows a trend from the bottom-left to the top-right, it indicates a positive correlation between the 'Attack'   and 'Defense' attributes.
 Pokemon with higher 'Attack' values also tend to have higher 'Defense' values.
 
 
### heatmap:

The correlation coefficient between 'attack' and 'attack' is 1.0, which indicates a perfect positive correlation. This is expected because a variable is always perfectly correlated with itself.

The correlation coefficient between 'attack' and 'defense' is 0.464479. This positive value indicates a moderate positive correlation between the 'attack' and 'defense' attributes. As 'attack' values increase, 'defense' values also tend to increase, and vice versa.
